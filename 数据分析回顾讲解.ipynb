### 数据分析

- numpy数组
    - 数组和列表之间的一个区别？
        - 数组中存储的数组元素的类型是统一，而列表的元素类型可以是任意。
    - 数组创建的方式3种：
        - 1.np.array(data)
        - 2.plt.imread(imgPath)
        - 3.routings系列的函数返回
            - np.random.xxx()
            - np.arange()
            - np.linespace()
    - 重要的属性
        - shape，ndim，size
    - 数组的索引
        - 取单行：arr[index]
        - 取多行：arr[[index1,index2...]]
        - 取元素：arr[index,col]
    - 切片：
        - 切行：arr[index1:index3]
        - 切列：arr[:,col1:col3]
        - 倒置：arr[::-1,::-1]
    - numpy数组主要是用来存储和运算数值型的数据
        - int,float
    - 如何给数组元素转换数据类型
        - 在创建数组的时候，可以使用一个dtype的参数指定数组元素类型
        - arr.dtype = 'xxx'
        - arr.astype
    - 变形reshape：
        - 可以将数组的维度进行任意的变换
            - 原则：变形前和变形后的数组容量得保持一致
    - 级联concatanate：可以将多个数组进行横向或者纵向的拼接。
        - 准则：只有相同维度的数组才可以发生级联。
    - 常见的聚合操作
        - max，min，mean，sum
        - 轴向axis：一般情况下0表示列，1表示行
    - 常见的统计函数：
        - std：标准差
        - var：方差
        - median：取中位数
    - 矩阵：
        - 矩阵的转置：.T(将行变成列，列变成行)
        - 矩阵的乘法：
            - 一个行向量乘以该向量的转置等于常数
        

- Series
    - 有两个常用的类：Series和DataFrame
    - Series就是一个一维的数据结构
    - Series是可以有显示和隐式索引
        - 显示索引的意义：可以增加数据的可读性
    - isnull，notnull：判断每一个元素是否为空
    - unique：去重
    - nunique：统计去重后的元素个数
    - value_counts：统计每一个不同元素出现的次数
    - Series的运算法则：
        - 只有索引一致的元素才可以进行算数运算，否则补空
    - Series的索引和切片的操作和列表是一致
    - Series可以使用布尔值充当索引去取值

- DataFrame
    - 组成元素：
        - 行索引
        - 列索引
        - 元素
    - 创建方式：
        - 可以使用numpy数组充当数据源：DataFrame(data=np.array([[]]))
        - 使用字典作为数据源（字典的key是作为df的列索引）
    - 重要的参数：
        - index：设置显示的行索引
        - columns：设置显示的列索引
    - 索引操作：
        - 单列：df[col]
        - 多列:df[[col1,col2。。。]]
        - 单行：df.loc[显示的行索引]，df.iloc[隐式的行索引]
        - 取元素：df.iloc[index,col]
    - 切片：
        - 切行：df[行切片]
        - 切列：df[:,列切片]
    - 财经数据接口包：tushare
    - 聚宽学习：https://www.cnblogs.com/bobo-zhang/category/1388806.html
    - df的持久化存储：df.to_xxx()
    - 将数据加载到df中：pd.read_xxx()
    - 时间类型转换：pd.to_datetime()
    - 单独将时间序列数据的年，月，日，周，进行单独的提取：
        - df[col].dt.day/month
    - 可以将某一列作为源数据的行索引：
        - set_index(col)
    - 布尔值是可以作为df的行列索引
    - 数据的重新取样：resample
    - 让Series的元素上下移动：shift(1)
    - df['close].rolling(5).mean()
- 数据清洗
    - 缺失值的处理
        - NAN和None的区别：数据类型不同，NAN是float类型和None是对象类型
        - dropna:可以直接将存有缺失数据的行或者列进行删除
        - fillna：可以使用指定值或者近邻值对缺失数据进行填充
            - 指定值：使用均值或者中位数
        - 一般情况下使用dropna，如果删除的成本高，则填充
    - 重复数据的清洗
        - drop_duplicates()可以将重复的行数据进行删除
    - 异常数据（离群点）
        - 制定一个判定异常数据的条件
- df的级联：pd.concat()可以将多个df进行横向或者纵向的拼接
    - 匹配级联&不匹配级联（会存在空值）
- 合并：pd.merge(how='outer',on,left_on,right_on)
- 合并和级联的区别：
    - 级联就是对表格做拼接
    - 合并是对应数据做汇总
- df.info()
    - 1.可以查看每一列的数据类型
    - 2.列中是否有缺失值
        - df.isnull().any(axis=0)
        - 想要计算列中空值的数量？
            - df[col].isnull().sum()
- describ():查看列的一些统计描述信息
    - 每一列的var，mean，max，min，中位数
- df实现条件查询：df.query('查询条件')
- df的排序：
    - sort_index
    - sort_value

- pandas的高级操作
    - 元素的替换replace
        - df.replace(to_replace=,value=)
        - 指定列的元素替换：df.replace(to_replace={col:value},value='xxx')
    - map映射
        - 在映射之前创建一个字典形式的映射关系表
        - map(dic)
        - 注意：map只可以被Series调用
    - map充当运算工具
        - 在充当运算工具前需要封装一个运算的函数
        - map(func)
        - apply也可以向map一样充当series的运算工具，且效率要高于map
    - 随机抽样：
        - df.take()：可以将df的行列打乱
            - take中只可以使用隐式形式的行列索引，无法是用显示索引
        - np.random.permutation(n):返回0-（n-1）的随机乱序序列
    - 分组：groupby
        - 参数：by充当分组条件，axis控制轴向
        - 分组的目的：是为了实现分组聚合
            - df.groupby(by=)['col'].mean()
        - 分组的高级聚合
            - 可以自己制定分组后的聚合函数
            - apply&transform
                - df.groupby(by=)['col'].apply(my_mean)
    - 透视表df.pivot_table：
        - 参数：
            - index：数据汇总的条件
            - value：指定列进行条件汇总
            - columns：可以对value进行一个细致分组
            - aggfunc：指定聚合的形式
    - df.apply:可以对df中的行或者列进行指定形式的操作
        - df.apply(func,axis=0)
    - df.applymap:可以对df中的每一个元素进行指定形式的运算
        - df.applymap(func)
    - df.groupby().agg([min,max]):可以对分组后的结果进行多种不同形式的聚合操作
