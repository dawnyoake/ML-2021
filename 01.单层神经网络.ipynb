```python
import numpy as np
import matplotlib.pyplot as plt
#  
X = np.array([[1,3,3],
              [1,4,3], 
              [1,1,1], 
              [1,2,1]])
#  
#
T = np.array([[1],
              [1], 
              [-1], 
              [-1]])
#31
#
W = np.random.random([3,1])
# 
lr = 0.1
# 
Y=0
#  
def train():
    #  W
    global W
    #  4 
    # Y (4,1)-4  1 
    Y = np.sign(np.dot(X,W))
    # T - Y  4  E(4,1)
    E=T-Y
    delta_W = lr * (X.T.dot(E)) 
    W = W + delta_W
for i in range(100): 
    # 
    train()
    #  
    print('epoch:',i + 1) 
    #  
    print('weights:',W)
    # 
    Y = np.sign(np.dot(X,W))
    # .all() Y  T  
    if(Y == T).all():
        print('Finished') 
        #  
        break
        
```
